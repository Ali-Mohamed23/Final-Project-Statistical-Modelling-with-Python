```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vancouver_bikes_df = pd.read_csv('vancouver_bikes_df')
    
yelp_combined_df = pd.read_csv('yelp_df')
```

```python
vancouver_bikes_df

van_df = vancouver_bikes_df.loc[[225, 67, 172, 232, 175]]

van_df['bike_station'] = [1, 2, 3, 4, 5]

van_df
```

```python
yelp_combined_df
```

```python
bikes_combined = pd.merge(van_df, yelp_combined_df, on='bike_station')

bikes_combined
```

```python
bikes_combined.to_csv('bikes_combined', index=False)
```

```python
sns.regplot(x='bike_station',y='rating', data=bikes_combined)
plt.show()
```

```python
bikes_combined.plot.bar(x = 'bike_station', y = 'free_bikes')
```

```python
sns.regplot(x='free_bikes',y='rating', data=bikes_combined)
plt.show()
```

# Database

Put all your results in an SQLite3 database (remember, SQLite stores its databases as files in your local machine - make sure to create your database in your project's data/ directory!)

```python
import sqlite3
import pandas as pd

#Create SQL database
conn = sqlite3.connect('city_bikes.db')

# Turn dataframe into pdfs to make available for SQL database
city_bikes_combined = pd.read_csv('bikes_combined')
fsq_df = pd.read_csv('fsq_df')

# Import dataframes into SQL database as tables
city_bikes_combined.to_sql(name='cityb_table', con=conn, if_exists='replace')
yelp_combined_df.to_sql(name='yelp_table', con=conn, if_exists='replace')
fsq_df.to_sql(name='fsq_table', con=conn, if_exists='replace')

# Close the database connection
conn.close()
```
